# Scrape 1st page:

```python
url="https://www.amazon.in/s?i=fashion&bbn=26925981031&rh=n%3A6648217031%2Cn%3A26925981031%2Cn%3A7459780031%2Cn%3A1953602031%2Cn%3A1968253031%2Cp_n_pct-off-with-tax%3A27060456031&s=apparel&dc&pf_rd_i=1953602031&pf_rd_m=A1VBAL9TL5WCBF&pf_rd_p=2578035d-442c-40a0-bb4c-a40b53448332&pf_rd_r=89T552YN82AMPD73MVDV&pf_rd_s=merchandised-search-14&qid=1681977080&rnid=2665398031&ref=sr_pg_1"
```

```python
req=requests.get(url,headers=HEADERS)
req
```

```python
soup=bs4.BeautifulSoup(req.text,'html.parser')
print(soup.prettify())
```

```python
br=[]
for brand in soup.find_all("span",class_='a-size-base-plus a-color-base'):
    br.append(brand.text)
    #print(brand)
    
len(br)
```

```python
br
```

```python
nm=[]

for name in soup.find_all("span",class_='a-size-base-plus a-color-base a-text-normal'):
    
    #print(name)
    nm.append(name.text)
    
len(nm)
```

```python
rt=[]

for rating in soup.find_all("span",class_='a-icon-alt'):
    
    #print(rating)
    rt.append(rating.text)
    
len(rt)
```

```python
pr=[]
for price in soup.find_all("span",class_='a-price-whole'):
    
    pr.append(price.text)

len(pr)
```

```python
df=pd.DataFrame({'Brand':br,'Name':nm,'Rating':rt,'Price':pr})
df
```

```python
df.to_excel('Amazon.xlsx')
```
